```python
import json
from bs4 import BeautifulSoup
```

## JSON

```python
obj = open("e_coli_core.json")
data = json.load(obj)
```

```python
reactions = data.get("reactions")
```

```python
for reaction in reactions:
    reaction_id = reaction['id']
    metabolites = reaction['metabolites']
    if len(metabolites) == 1:
        print(f"{reaction_id} :{[x for x,y in metabolites.items()][0]} <=>\n")
        
    else:
        lower = reaction['lower_bound']
        upper = reaction['upper_bound']
        if lower == 0:
            arrow = "=>"
        elif upper == 0:
            arrow = "<="
        else:
            arrow = "<=>"
        l = [str((y)) + ' ' + x for x,y in metabolites.items() if y > 0]
        l = ' + '.join(l)
        r = [str(-y) + ' ' +  x for x,y in metabolites.items() if y < 0]
        r = ' + '.join(r)
        print(f"{reaction_id}: {l} {arrow} {r}\n")
```

## XML

```python
file = open("e_coli_core.xml")
soup = BeautifulSoup(file, "html.parser")
```

```python
reactions = soup.sbml.model.listofreactions.findAll('reaction')
```

```python
for reaction in reactions:
    reversible = reaction['reversible']
    if reversible == 'false':
        arrow = "=>"
    else:
        arrow = '<=>'
    reaction_id = reaction['id'][2:]
    
    if reaction.listofreactants != None:
        reactants = reaction.listofreactants.find_all('speciesreference')
        r = ' + '.join([str(reactants[i]['stoichiometry']) + ' ' + str(reactants[i]['species'][2:]) for i in range(len(reactants))])
    else:
        r = ''
    
    if reaction.listofproducts != None:
        products = reaction.listofproducts.find_all('speciesreference')
        p = ' + '.join([str(products[i]['stoichiometry']) + ' ' + str(products[i]['species'][2:]) for i in range(len(products))])
    else:
        p = ''
        
    print(f'{reaction_id}: {r} {arrow} {p}\n')
```
